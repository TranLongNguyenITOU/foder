```python
import pandas as pd
import matplotlib.pyplot as plt
# Đường dẫn đến file CSV
file_path = 'LamPhat.csv'
# Đọc file CSV và lưu vào một DataFrame
df = pd.read_csv(file_path)
# Hiển thị nội dung của DataFrame
print(df)
# Giả sử bạn muốn vẽ biểu đồ đường với trục x là cột 'year' và trục y là cột 'value'.
x = df['Time']
y = df['Value']

# Vẽ biểu đồ đường
plt.plot(x, y)

# Thêm tiêu đề và nhãn trục
plt.title('Biểu đồ đường cho tỷ lệ lạm phát qua các năm')
plt.xlabel('Năm')
plt.ylabel('Tỷ lệ (%)')

# Hiển thị biểu đồ
plt.show()
```
