```python
import pandas as pd
import matplotlib.pyplot as plt
# Đường dẫn đến file CSV
df = pd.read_csv('GDP.csv')
# Đọc file CSV và lưu vào một DataFrame
print(df)
# Giả sử bạn muốn vẽ biểu đồ đường với trục x là cột 'year' và trục y là cột 'value'.
x = df['Time']
y = df['Value']
# Vẽ biểu đồ đường
plt.plot(x, y)

# Thêm tiêu đề và nhãn trục
plt.title('Biểu đồ đường cho tốc độ tăng trưởng GDP')
plt.xlabel('Năm')
plt.ylabel('Giá trị')

# Hiển thị biểu đồ
plt.show()
```
