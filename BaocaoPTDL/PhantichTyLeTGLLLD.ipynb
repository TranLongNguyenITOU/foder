```python
import pandas as pd
import matplotlib.pyplot as plt
# Đường dẫn đến file CSV
file_path = 'TyLeTGLLLaoDong.csv'
# Đọc file CSV và lưu vào một DataFrame
df = pd.read_csv(file_path)
col = ['ref_area.label','indicator.label','source.label', 'sex.label', 'classif1.label', 'classif2.label', 'obs_status.label', 'note_classif.label','note_classif.label','note_source.label','note_indicator.label']
df.drop(columns=col, inplace=True)
df = df.rename(columns={'obs_value': 'value'})
# Hiển thị nội dung của DataFrame
print(df)
# Giả sử bạn muốn vẽ biểu đồ đường với trục x là cột 'year' và trục y là cột 'value'.
x = df['time']
y = df['value']

# Vẽ biểu đồ đường
plt.plot(x, y)

# Thêm tiêu đề và nhãn trục
plt.title('Biểu đồ đường cho Tỷ Lệ Tham Gia Lực Lượng Lao Động')
plt.xlabel('Năm')
plt.ylabel('Tỷ lệ (%)')

# Hiển thị biểu đồ
plt.show()
df.to_csv('TyLeTGLLLDdaxuly.csv', encoding='utf-8', index=False, header=True)
```
