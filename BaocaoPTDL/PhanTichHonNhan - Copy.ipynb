```python
import pandas as pd
import matplotlib.pyplot as plt
# Đường dẫn đến file CSV
file_path = 'TNtheoHonNhan.csv'

# Đọc file CSV và lưu vào một DataFrame
df = pd.read_csv(file_path)
df = df.rename(columns={'ref_area.label': 'Country', 'obs_value': 'Value', 'classif2.label':'Marital status'})
col = ['sex.label','indicator.label', 'time', 'source.label', 'obs_status.label', 'note_classif.label', 'note_indicator.label', 'note_source.label','classif1.label' ]
df.drop(columns=col, inplace=True)
mapping_dict = {
    'Marital status (Detailed): Single': 'Single',
    'Marital status (Detailed): Married': 'Married'
}
df.replace(mapping_dict, inplace=True)
print(df)
# Tính tổng giá trị (Value) cho mỗi trạng thái hôn nhân
grouped_df = df.groupby('Marital status')['Value'].sum()

# Tính toán phần trăm cho mỗi trạng thái hôn nhân
total_value = grouped_df.sum()
percentages = (grouped_df / total_value) * 100

# Vẽ biểu đồ tròn
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=percentages.index, autopct='%1.1f%%', startangle=140)
plt.title('Percentage of Marital Status')
plt.axis('equal')  # Đảm bảo biểu đồ tròn
plt.show()

```
